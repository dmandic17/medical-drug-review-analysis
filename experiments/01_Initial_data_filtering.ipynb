# Initial Data Filtering

**In this notebook, we filter the data such that the new data set cotains only the medical conditions that are mentioned in more than 500 rows.**

Note: original data set - "drugsComTrain_raw.csv"; new data set - "drugsComTrain_top_conditions.csv";

```python
import pandas as pd
data = pd.read_csv("../data/drugsComTrain_raw.csv")
data.info()
```

```python
i = (50240)
print(data.iloc[i])
print(data["review"][i])
```

```python
data[data["condition"].str.contains("schizophrenia", case=False)==True]["review"].iloc[35]
```

```python
data.groupby("condition").count().sort_values("uniqueID", ascending=False)[10:70]
```

```python
#data.groupby("drugName").count().sort_values("uniqueID", ascending=False)[0:50]
condition_count_dict = data["condition"].value_counts()
data["condition_count"] = data["condition"].map(condition_count_dict)
data_top_conditions = data.drop(data[data["condition_count"]<500].index)
data_top_conditions
```

```python
data_top_conditions.to_csv("../data/drugsComTrain_top_conditions.csv")
```
